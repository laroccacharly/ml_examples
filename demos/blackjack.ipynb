```python
import gym
from src.agent import MonteAgent, QlearnAgent
from src.utils import HyperFitter
import matplotlib.pyplot as plt
```

Let's compare Monte Carlo and Q-Learning on OpenAI toy env Blackjack-v0 


```python
env = gym.make('Blackjack-v0') 
monte_agent = MonteAgent(env=env)
qlearn_agent = QlearnAgent(env=env)
epsilons = [0.005, 0.01, 0.05, 0.1, 0.2, 0.9]
nb_episodess = [1000]
space = {'epsilon': epsilons, 'nb_episodes': nb_episodess}
nb_iter = 10 
monte_hf = HyperFitter(agent=monte_agent, space=space, nb_iter=nb_iter)
qlearn_hf = HyperFitter(agent=qlearn_agent, space=space, nb_iter=nb_iter)
```

```python
monte_hf.fit()
monte_hf.plot(x_axis_name='epsilon', compare_on='nb_episodes')
```

```python
qlearn_hf.fit()
plt = qlearn_hf.plot(x_axis_name='epsilon', compare_on='nb_episodes')
```
