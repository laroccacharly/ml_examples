```python
%matplotlib inline

import gym
from src.agent import MonteAgent
from src.utils import HyperFitter
import matplotlib.pyplot as plt

env = gym.make('Blackjack-v0')
agent = MonteAgent(env=env, nb_episodes=10000)
epsilons = [0, 0.01, 0.05, 0.1, 0.2, 0.3, 0.9]
space = {'epsilon': epsilons}
hf = HyperFitter(agent=agent, space=space)
hf.fit()

plt.plot(epsilons, hf.scores)
```

```python
from src.agent import QlearnAgent
env = gym.make('FrozenLake-v0')
agent = QlearnAgent(env=env, nb_episodes=1000)
hf_qlearn = HyperFitter(agent=agent, space=space)
hf_qlearn.fit()

plt.plot(epsilons, hf_qlearn.scores)
```
